## What are SKILL.md Files?

SKILL.md files are **instruction documents** that teach AI agents (running in IDEs) about your project, coding patterns, and domain knowledge. They help agents understand how to work better with your codebase.

### Supported IDEs
- **VS Code** - with GitHub Copilot
- **Cursor** - AI-first code editor
- **Aider** - command-line AI pair programmer
- **Antigraivity** - AI-powered IDE

### Purpose of SKILL.md
- Improve code understanding
- Define project conventions and patterns
- Guide agent behavior with rules
- Document workflows and best practices
- Enable multi-step task execution

### Key Components
SKILL.md files use **YAML frontmatter** for metadata:

```yaml
---
name: Your Skill Name
description: Brief description
triggers: ["keyword1", "keyword2"]
applyTo: ["*.py", "src/**"]
---
```

Official Documentation:
- [VS Code Agent Customization](https://code.visualstudio.com/docs/copilot/overview)
- [Cursor Settings](https://docs.cursor.sh/)
- [Aider Configuration](https://aider.chat/)

## YAML Frontmatter Syntax

SKILL.md files start with YAML frontmatter enclosed in triple dashes. This metadata tells the IDE agent how to apply the skill.

### Common Frontmatter Fields

```yaml
---
name: string              # Skill name (required)
description: string       # What this skill does
triggers: [string]        # Keywords that activate this skill
applyTo: [glob]          # File patterns to apply to
priority: number         # Execution priority (higher runs first)
version: string          # Skill version
---
```

### Example Frontmatter

```yaml
---
name: Python Testing Skill
description: Improves Python test generation and debugging
triggers: ["test", "unittest", "pytest"]
applyTo: ["**/*.py", "tests/**"]
priority: 10
version: 1.0
---
```

The content after the closing `---` provides the instructions that guide the agent's behavior.

````python
# SKILL.md example as a Python string (for demonstration)
skill_example = """---
name: FastAPI API Patterns
description: Teaches the agent FastAPI best practices and patterns
triggers: ["fastapi", "api", "endpoint"]
applyTo: ["**/*.py"]
priority: 5
version: 1.0
---

## FastAPI API Patterns Skill

You are helping developers build FastAPI applications. When working with FastAPI code:

### Rules
1. Always use type hints for request/response models
2. Use Pydantic BaseModel for data validation
3. Include docstrings for all endpoints
4. Use status codes correctly (200, 201, 400, 404, 500)

### Common Patterns
- Create POST endpoints for data mutation
- Use GET endpoints for data retrieval
- Use PUT/PATCH for updates
- Always validate input with Pydantic models

### Tools Available
- pytest for testing
- httpx for client requests
- TestClient from fastapi.testclient
"""

print("SKILL.md Example:")
print("=" * 50)
print(skill_example)
````

## Rules and Workflows in IDE Agents

Skills define **rules** that guide agent behavior and **workflows** for complex multi-step tasks.

### Rules
Rules are constraints and guidelines that ensure the agent follows best practices:

```
You are working with Python code. Rules:
1. Always use type hints
2. Write docstrings for public functions
3. Follow PEP 8 naming conventions
4. Use list comprehensions over loops when appropriate
```

### Workflows
Workflows orchestrate multi-step processes:

```
Workflow: Writing Unit Tests
1. Analyze the function to test
2. Identify edge cases
3. Generate test cases
4. Create test file
5. Run tests
6. Report results
```

### Tool Calling
Within SKILL.md, agents can reference available tools:
- **Code search**: Find relevant code patterns
- **File operations**: Read/write files
- **Execution**: Run tests, linters, scripts
- **Context retrieval**: Get project documentation

## Practical SKILL.md Example

Let's create a complete SKILL.md example for a Python FastAPI project that demonstrates rules, workflows, and best practices.

````python
# Complete SKILL.md Example with Workflows
complete_skill = """---
name: Python Testing and Debugging
description: Comprehensive skill for testing and debugging Python applications
triggers: ["test", "debug", "unittest", "pytest", "error"]
applyTo: ["**/*.py", "tests/**/*"]
priority: 10
version: 1.0
---

# Python Testing and Debugging Skill

## Overview
This skill teaches the agent best practices for writing tests, debugging code, and ensuring code quality in Python projects.

## Core Rules

1. **Type Safety**: Always use type hints (Python 3.7+)
   - Use `from typing import Optional, List, Dict`
   - Annotate function parameters and return types
   - Use TypedDict for complex dictionaries

2. **Testing Standards**:
   - Write unit tests for all public functions
   - Use pytest as the primary test framework
   - Aim for 80%+ code coverage
   - Name test functions with `test_` prefix
   - Use fixtures for setup/teardown

3. **Error Handling**:
   - Catch specific exceptions, not bare Exception
   - Provide meaningful error messages
   - Use logging instead of print statements
   - Log errors at appropriate levels (DEBUG, INFO, WARNING, ERROR)

4. **Code Quality**:
   - Follow PEP 8 style guide
   - Use meaningful variable names
   - Keep functions under 50 lines
   - Document complex logic with comments

## Multi-step Workflows

### Workflow: Fix and Test
1. Analyze error message and stack trace
2. Locate the problematic code
3. Identify root cause
4. Create fix
5. Write test case to prevent regression
6. Run full test suite
7. Verify all tests pass

### Workflow: Add New Feature with Tests
1. Read existing code patterns
2. Plan implementation based on conventions
3. Write test first (TDD approach)
4. Implement feature
5. Run tests until they pass
6. Review code quality
---

## Available Tools
- `pytest` - Run tests and check coverage
- `pylint` / `flake8` - Code quality checks
- `mypy` - Static type checking
- `debugpy` - Python debugging
"""

print(complete_skill)
print("\n" + "=" * 70)
print("This SKILL.md would be saved as a file in your project root")
print("IDE agents would automatically discover and use this skill!")
````

## Step-by-Step: Creating a SKILL.md for Your Project

### Step 1: Create the File
```bash
# In your project root directory
touch SKILL.md
touch .cursorrules  # Optional, for Cursor IDE
```

### Step 2: Add YAML Frontmatter
```yaml
---
name: My Project Guide
description: AI agent instructions for this project
triggers: ["help", "code", "test"]
applyTo: ["src/**/*.py"]
priority: 10
version: 1.0
---
```

### Step 3: Document Your Rules
Write clear, concise instructions that help the agent understand:
- Your coding standards
- Project structure
- Common patterns
- Testing requirements

### Step 4: Define Workflows
Create step-by-step processes for:
- Code review procedures
- Testing workflows
- Debugging approaches
- Feature implementation

````python
# Create a simple SKILL.md file that students can use
sample_skill_content = """---
name: Student Project Guide
description: SKILL.md example for AI-enabled IDEs
triggers: ["help", "fix", "test", "debug"]
applyTo: ["**/*.py"]
priority: 10
version: 1.0
---

# Student Project AI Assistant Guide

## Project Overview
This is a training project using FastAPI for building APIs.

## Core Rules for AI Agent

### 1. Code Style
- Use Python 3.9+ type hints
- Follow PEP 8 naming conventions
- Maximum line length: 100 characters
- Use docstrings for all public functions

### 2. Testing Requirements
- Write unit tests for all functions
- Use pytest as test framework
- Test files in tests/ directory
- Aim for 80%+ coverage

### 3. Error Handling
- Never use bare except clauses
- Provide descriptive error messages
- Use logging module for debug info
- Include stack traces in production logs

### 4. Git Commits
- Write clear commit messages
- One feature per commit
- Reference issue numbers when applicable

## Common Workflows

### Workflow: Fix a Bug
1. Read error message carefully
2. Find the problematic function
3. Add debug logging
4. Write a failing test that reproduces the bug
5. Fix the bug
6. Verify test passes
7. Clean up debug code
8. Create a commit with message "fix: [description]"

### Workflow: Add a New Feature
1. Create a new function with type hints
2. Add comprehensive docstring
3. Write test cases (before implementation if using TDD)
4. Implement the feature
5. Run full test suite
6. Do code review check
7. Commit with message "feat: [description]"

## Tools Available
- pytest: Run tests
- pylint: Code quality analysis
- mypy: Type checking
- black: Code formatting
"""

# Save to file
with open("SKILL_EXAMPLE.md", "w") as f:
    f.write(sample_skill_content)

print("✅ Sample SKILL.md created successfully!")
print("\nFile saved as: SKILL_EXAMPLE.md")
print("\nTo use this in your project:")
print("  1. Copy content to SKILL.md in project root")
print("  2. Customize rules and workflows for your project")
print("  3. Restart your IDE")
print("  4. The agent will apply these rules automatically")
````

## Converting SKILL.md for Different IDEs

Each IDE has slightly different directory structures and file naming conventions:

### Antigraivity (Recommended approach)
```bash
# Create the skills directory structure
mkdir -p .agents/skills/my-skill

# Create SKILL.md inside the skill folder
touch .agents/skills/my-skill/SKILL.md

# Directory structure:
# <project-root>/
# └── .agents/
#     └── skills/
#         └── my-skill/
#             └── SKILL.md
```

### Cursor IDE
```bash
# Cursor uses .cursorrules file (same format, different file)
cp SKILL.md .cursorrules
```

### VS Code
```bash
# VS Code uses .copilot-instructions.md or SKILL.md
# Place in project root
# VS Code will automatically detect it
```

### Aider
```bash
# Aider reads from configuration
# Add system prompt from SKILL.md to .aider.conf.yaml
# Or use with: aider --system-prompt-file SKILL.md
```

**Key Point**: The SKILL.md **content** is compatible across all IDEs, but the **directory location** differs by IDE.
