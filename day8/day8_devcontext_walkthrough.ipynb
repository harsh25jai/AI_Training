## 1) What is DevContext?

DevContext is a Model Context Protocol (MCP) server focused on continuous developer context awareness. It maintains project-specific context over time, records code and conversation state, and provides retrieval tools like:

- `initialize_conversation_context`
- `update_conversation_context`
- `retrieve_relevant_context`
- `record_milestone_context`
- `finalize_conversation_context`

DevContext aims to reduce cognitive load by surfacing code snippets, design decisions, and relevant artifacts automatically while you develop.

## 2) Why this is useful for developers

- Keeps large, active codebases searchable in context (not just file text).
- Helps AI assistants stay on-topic by injecting project structure, function signatures, and prior decisions.
- Enables prompter workflow: ask about a function/database schema and get accurate context.
- Collects milestones/changes so automated agents can plan reliably.

> Result: more precise code-aware AI output and fewer repeated explanations.

## 3) Key features from GitHub README (devcontext)

- Node.js + TursoDB setup (recommended) with minimal external dependencies.
- Single-project DB per instance to avoid cross-project contamination.
- Non-vector retrieval using keyword graphs + metadata.
- Cursor Rules support for structured, persistent session management.
- MCP JSON configuration with `mcp.json` integration for Cursor (or compatible IDE).

## 4) Quick dev setup (macOS + VS Code) – Online or Offline Database

### Option A: Online Database (TursoDB Cloud) – Recommended for Teams

#### Prerequisites

1. Node.js 18+ installed.
2. VS Code with GitHub Copilot extension.
3. TursoDB account (free tier available at https://turso.tech).

#### Step-by-step setup

**1. Install Turso CLI**

```bash
# Install via Homebrew (macOS)
brew install tursodatabase/tap/turso

# Verify installation
turso --version
```

**2. Authenticate with Turso**

```bash
# For first-time users, sign up:
turso auth signup

# Or login if you already have an account:
turso auth login

# Verify authentication
turso auth whoami
```

**3. Create a Database**

```bash
# Create a new database named 'devcontext'
turso db create devcontext

# Get the database URL (libsql format)
turso db show devcontext --url
# Output: libsql://your-database-name-random-chars.turso.io
```

**4. Create Database Access Token**

```bash
# Generate an access token for the database
turso db tokens create devcontext

# This outputs a token like: v2_xyz... (keep it safe)
```

Save both the **database URL** and **token** for the next step.

**5. Configure DevContext for VS Code**

Create or edit `.vscode/settings.json` in your project directory:

```json
{
  "modelContext.servers": {
    "devcontext": {
      "command": "npx",
      "args": ["-y", "devcontext@latest"],
      "enabled": true,
      "env": {
        "TURSO_DATABASE_URL": "libsql://your-database-name-random-chars.turso.io",
        "TURSO_AUTH_TOKEN": "v2_xyz..."
      }
    }
  }
}
```

**6. Start DevContext**

```bash
cd ~/Desktop/Projects/Ace_Turtle/ai_training
npx -y devcontext@latest
```

---

### Option B: Offline Database (Local SQLite) – Fast, No Setup

#### Prerequisites

1. Node.js 18+ installed.
2. VS Code with GitHub Copilot extension.

#### Step-by-step setup

**1. Configure DevContext for VS Code (SQLite)**

Create `.vscode/settings.json` in your project directory:

```json
{
  "modelContext.servers": {
    "devcontext": {
      "command": "npx",
      "args": ["-y", "devcontext@latest"],
      "enabled": true,
      "env": {
        "DATABASE_URL": "sqlite:./devcontext.db"
      }
    }
  }
}
```

This creates a local `devcontext.db` file in your project root.

**2. Start DevContext**

```bash
cd ~/Desktop/Projects/Ace_Turtle/ai_training
npx -y devcontext@latest
```

The SQLite database initializes automatically on first run.

### Chat Workflow (Same for Both Options)

1. **Initialize Conversation**:
   - Open VS Code Chat (Copilot or built-in AI).
   - Ask: *"Call initialize_conversation_context to set up project context."*
   - AI returns a conversation ID.

2. **Make Code Changes**:
   - Edit your code in the project.
   - Ask: *"Call update_conversation_context with my recent code changes."*
   - AI syncs changes into DevContext.

3. **Retrieve Relevant Context** (as needed):
   - Ask: *"Call retrieve_relevant_context for [your question]"*.
   - Example: *"Call retrieve_relevant_context for how the authentication system works."*
   - AI retrieves project-specific code snippets.

4. **Record Milestones** (optional):
   - After completing a feature: *"Call record_milestone_context for [feature name]."*
   - AI logs the milestone for future reference.

5. **Finalize Conversation** (end of session):
   - Before closing: *"Call finalize_conversation_context."*
   - AI extracts learnings and suggests next steps.

## 6) DevContext + VS Code + GitHub Copilot integration

### Why This Combination Works

**VS Code + GitHub Copilot** provides a native AI assistant environment where you can:
- Write code naturally with AI suggestions.
- Ask questions in the Chat view while coding.
- Access MCP tools (like DevContext) seamlessly.

**DevContext** enhances this by:
- Maintaining code-aware conversation context.
- Retrieving project-specific snippets on demand.
- Recording development milestones and patterns.
- Enabling truly informed AI responses about your codebase.

### Integration Pattern

```
VS Code Chat
    ↓
GitHub Copilot (or compatible AI)
    ↓
DevContext MCP Tools (5 core functions)
    ↓
TursoDB (project context storage)
```

### Example Workflow in VS Code

**Scenario**: You're building a feature and need context about existing authentication.

1. **In VS Code Chat**: *"I need to add a refresh token mechanism. What's the current auth structure?"*
2. **Copilot internally**:
   - Calls `retrieve_relevant_context` with query: "authentication structure"
   - Receives code snippets from DevContext about auth modules.
3. **Copilot responds** with accurate, project-specific code suggestions.
4. **You implement** and later ask: *"Record that we added refresh token flow."*
5. **Copilot calls** `record_milestone_context` for future reference.

### Comparison: Before vs. After DevContext

| Aspect | Without DevContext | With DevContext |
|--------|-------------------|-----------------|
| **Context retrieval** | Copy-paste file contents | Automatic, precise snippets |
| **Relevance** | Generic AI knowledge | Project-specific insights |
| **Consistency** | Repeat explanations often | Context persists across chats |
| **Tracking* | Manual | Automatic milestones logged |

## 8) VS Code Troubleshooting & Quick Reference

### Database Selection Guide

| Symptom | Likely Issue | Solution |
|---------|--------------|----------|
| **Can't log into TursoDB** | Network or credentials issue | Use **Option B (SQLite)** instead |
| **Want to work offline** | TursoDB requires internet | Switch to **Option B (SQLite)** |
| **Need to share context with team** | SQLite is local-only | Switch to **Option A (TursoDB)** |

---

### Common Issues & Solutions

| Issue | Cause | Solution |
|-------|-------|----------|
| **"MCP server not found"** | Config missing or wrong | Verify `.vscode/settings.json`; reload VS Code |
| **"Connection refused"** | DevContext not running | Start: `npx -y devcontext@latest` |
| **"Invalid TursoDB credentials"** | Wrong URL/token | Re-run: `turso db show devcontext --url` & `turso db tokens create devcontext` |
| **"database.db file not created"** | SQLite not initialized | Restart DevContext; check disk permissions |
| **"No MCP tools available"** | VS Code not reloaded | Reload: `Cmd+Shift+P` → "Reload Window" |
| **"Conversation ID not recognized"** | Tool called out of order | Call `initialize_conversation_context` first |
| **"Permission denied" (SQLite)** | Disk write issue | Ensure `devcontext.db` is writable; check `.gitignore` |

### CLI Commands Reference (Based on Official Turso Docs)

**Installation & Authentication (Option A)**
```bash
brew install tursodatabase/tap/turso       # Install Turso CLI
turso auth signup                          # Sign up for new account
turso auth login                           # Login to existing account
turso auth whoami                          # Check current user
turso auth token                           # Get current API token
```

**Database Management (Option A)**
```bash
turso db create devcontext                 # Create new database
turso db list                              # List all databases
turso db show devcontext                   # Show database info
turso db show devcontext --url             # Show database URL (libsql format)
turso db show devcontext --http-url        # Show HTTP API URL
turso db delete devcontext                 # Delete database
```

**Database Tokens (Option A)**
```bash
turso db tokens create devcontext          # Create database token (default: full access, never expires)
turso db tokens create devcontext --read-only     # Create read-only token
turso db tokens create devcontext --expiration 7d # Token expires in 7 days
turso db tokens list devcontext            # List all tokens for database
turso db tokens invalidate devcontext <token-hash>  # Revoke a token
```

**DevContext (Both Options)**
```bash
npx -y devcontext@latest                   # Start MCP server
npm list devcontext                        # Check installed version (if local)
```

### Pro Tips for VS Code + DevContext

1. **Keep DevContext running in a dedicated terminal tab** while developing.
2. **Use workspace settings** (`.vscode/settings.json`) to share config with team (Option A only).
3. **For Option B (SQLite), add to `.gitignore`**: `echo "devcontext.db" >> .gitignore`
4. **Pin your VS Code Chat window** to prevent accidental closure.
5. **Test MCP availability** by typing tool names in chat: `initialize_conversation_context` should auto-complete.
6. **Monitor DevContext output** for performance warnings or indexing progress (first run may take time).
7. **Backup SQLite database periodically** (Option B): `cp devcontext.db devcontext.db.backup`
8. **Database URL format**: Should be `libsql://database-name.turso.io` (not `https://`)
9. **Token security** (Option A): Store tokens in `.gitignore` or environment variables for production
