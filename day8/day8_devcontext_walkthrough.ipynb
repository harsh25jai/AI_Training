## 1) What is DevContext?

DevContext is a Model Context Protocol (MCP) server focused on continuous developer context awareness. It maintains project-specific context over time, records code and conversation state, and provides retrieval tools like:

- `initialize_conversation_context`
- `update_conversation_context`
- `retrieve_relevant_context`
- `record_milestone_context`
- `finalize_conversation_context`

DevContext aims to reduce cognitive load by surfacing code snippets, design decisions, and relevant artifacts automatically while you develop.

## 2) Why this is useful for developers

- Keeps large, active codebases searchable in context (not just file text).
- Helps AI assistants stay on-topic by injecting project structure, function signatures, and prior decisions.
- Enables prompter workflow: ask about a function/database schema and get accurate context.
- Collects milestones/changes so automated agents can plan reliably.

> Result: more precise code-aware AI output and fewer repeated explanations.

## 3) Key features from GitHub README (devcontext)

- Node.js + TursoDB setup (recommended) with minimal external dependencies.
- Single-project DB per instance to avoid cross-project contamination.
- Non-vector retrieval using keyword graphs + metadata.
- Cursor Rules support for structured, persistent session management.
- MCP JSON configuration with `mcp.json` integration for Cursor (or compatible IDE).

## 4) Quick dev setup (macOS + VS Code)

### Prerequisites

1. Node.js 18+ installed.
2. VS Code with GitHub Copilot extension (or other MCP-compatible AI assistant).
3. TursoDB account and credentials.

### Step-by-step setup

#### 1. Set up TursoDB

```bash
# Install Turso CLI
curl -sSfL https://get.turso.tech/install.sh | bash

# Authenticate
turso auth login

# Create database
turso db create devcontext

# Get database URL
turso db show devcontext --url

# Create auth token
turso db tokens create devcontext
```

Save both the **database URL** and **auth token** for the next step.

#### 2. Configure DevContext for VS Code

Create `.vscode/settings.json` in your project directory (or edit existing):

```json
{
  "modelContext.servers": {
    "devcontext": {
      "command": "npx",
      "args": ["-y", "devcontext@latest"],
      "enabled": true,
      "env": {
        "TURSO_DATABASE_URL": "libsql://your-project-db-name.turso.io",
        "TURSO_AUTH_TOKEN": "your_turso_auth_token_here"
      }
    }
  }
}
```

Alternatively, if using workspace config, update `.vscode/settings.json` with the above.

#### 3. Verify Installation

From your project root, start DevContext manually:

```bash
cd ~/Desktop/Projects/Ace_Turtle/ai_training
npx -y devcontext@latest
```

You should see output indicating the MCP server is running and listening for connections.

#### 4. Restart VS Code

Close and reopen VS Code, or reload the window (`Cmd+Shift+P` → "Developer: Reload Window").

#### 5. Test with GitHub Copilot or VS Code's Built-in AI

Open a chat in VS Code and ask a code-related question. DevContext MCP tools should now be available:
- `initialize_conversation_context`
- `update_conversation_context`
- `retrieve_relevant_context`
- `record_milestone_context`
- `finalize_conversation_context`

## 5) Suggested learning flow (for VS Code)

### Workflow

1. **Setup DevContext** (from section 4 above):
   - Create TursoDB database and get credentials.
   - Add MCP configuration to `.vscode/settings.json`.
   - Start DevContext with `npx -y devcontext@latest`.
   - Restart VS Code.

2. **Initialize Conversation**:
   - Open VS Code Chat (Copilot or built-in AI).
   - Ask your AI assistant: *"Call initialize_conversation_context to set up project context."*
   - The AI will invoke the MCP tool and return a conversation ID.

3. **Make Code Changes**:
   - Edit your code in the project.
   - Ask your AI: *"Call update_conversation_context with my recent code changes."*
   - AI will sync your changes into DevContext.

4. **Retrieve Relevant Context** (as needed):
   - Ask your AI: *"Call retrieve_relevant_context for _____ [specific question]"*.
   - Example: *"Call retrieve_relevant_context for how the authentication system works."*
   - AI will retrieve code snippets and explanations.

5. **Record Milestones** (optional but useful):
   - After completing a feature: *"Call record_milestone_context for [feature name]."*
   - AI will log the milestone for future reference.

6. **Finalize Conversation** (end of session):
   - Before closing VS Code or ending conversation: *"Call finalize_conversation_context."*
   - AI extracts learnings and suggests next steps.

## 6) DevContext + VS Code + GitHub Copilot integration

### Why This Combination Works

**VS Code + GitHub Copilot** provides a native AI assistant environment where you can:
- Write code naturally with AI suggestions.
- Ask questions in the Chat view while coding.
- Access MCP tools (like DevContext) seamlessly.

**DevContext** enhances this by:
- Maintaining code-aware conversation context.
- Retrieving project-specific snippets on demand.
- Recording development milestones and patterns.
- Enabling truly informed AI responses about your codebase.

### Integration Pattern

```
VS Code Chat
    ↓
GitHub Copilot (or compatible AI)
    ↓
DevContext MCP Tools (5 core functions)
    ↓
TursoDB (project context storage)
```

### Example Workflow in VS Code

**Scenario**: You're building a feature and need context about existing authentication.

1. **In VS Code Chat**: *"I need to add a refresh token mechanism. What's the current auth structure?"*
2. **Copilot internally**:
   - Calls `retrieve_relevant_context` with query: "authentication structure"
   - Receives code snippets from DevContext about auth modules.
3. **Copilot responds** with accurate, project-specific code suggestions.
4. **You implement** and later ask: *"Record that we added refresh token flow."*
5. **Copilot calls** `record_milestone_context` for future reference.

### Comparison: Before vs. After DevContext

| Aspect | Without DevContext | With DevContext |
|--------|-------------------|-----------------|
| **Context retrieval** | Copy-paste file contents | Automatic, precise snippets |
| **Relevance** | Generic AI knowledge | Project-specific insights |
| **Consistency** | Repeat explanations often | Context persists across chats |
| **Tracking* | Manual | Automatic milestones logged |

## 8) VS Code Troubleshooting & Quick Reference

### Common Issues

| Issue | Solution |
|-------|----------|
| **"MCP server not found"** | Verify `.vscode/settings.json` has correct config; restart VS Code. |
| **"Connection refused"** | Ensure DevContext is running: `npx -y devcontext@latest` from project root. |
| **"Invalid credentials"** | Double-check `TURSO_DATABASE_URL` and `TURSO_AUTH_TOKEN` in settings. |
| **"No MCP tools available"** | Reload VS Code window (`Cmd+Shift+P` → "Reload Window"). |
| **"Conversation ID not recognized"** | Make sure `initialize_conversation_context` was called first. |

### CLI Commands Reference

```bash
# Start DevContext
npx -y devcontext@latest

# View TursoDB database details
turso db show devcontext --url

# Generate new auth token (if needed)
turso db tokens create devcontext

# List all Turso databases
turso db list
```

### File Locations

- **Project MCP Config**: `.vscode/settings.json` (workspace-specific)
- **DevContext Server**: Runs on default MCP port (check console output)
- **Database**: Hosted on TursoDB (cloud-based, no local files)
- **Project Root**: Must be where you run `npx devcontext@latest`

### Pro Tips for VS Code

1. **Keep DevContext running in a terminal tab** while developing.
2. **Use workspace settings** (`.vscode/settings.json`) to share config with team.
3. **Pin your chat window** to prevent accidental closure.
4. **Test MCP availability** by typing tool names in chat: `initialize_conversation_context` should auto-complete.
5. **Save environment variables** in `.env.local` if using local setup (GitIgnore it).
