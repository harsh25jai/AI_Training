## 1) What is DevContext?

DevContext is a Model Context Protocol (MCP) server focused on continuous developer context awareness. It maintains project-specific context over time, records code and conversation state, and provides retrieval tools like:

- `initialize_conversation_context`
- `update_conversation_context`
- `retrieve_relevant_context`
- `record_milestone_context`
- `finalize_conversation_context`

DevContext aims to reduce cognitive load by surfacing code snippets, design decisions, and relevant artifacts automatically while you develop.

## 2) Why this is useful for developers

- Keeps large, active codebases searchable in context (not just file text).
- Helps AI assistants stay on-topic by injecting project structure, function signatures, and prior decisions.
- Enables prompter workflow: ask about a function/database schema and get accurate context.
- Collects milestones/changes so automated agents can plan reliably.

> Result: more precise code-aware AI output and fewer repeated explanations.

## 3) Key features from GitHub README (devcontext)

- Node.js + TursoDB setup (recommended) with minimal external dependencies.
- Single-project DB per instance to avoid cross-project contamination.
- Non-vector retrieval using keyword graphs + metadata.
- Cursor Rules support for structured, persistent session management.
- MCP JSON configuration with `mcp.json` integration for Cursor (or compatible IDE).

## 4) Quick dev setup (macOS)

1. Node.js 18+ installed.
2. Setup TursoDB and token.
3. Create `.cursor/mcp.json`:

```json
{
  "mcpServers": {
    "devcontext": {
      "command": "npx",
      "args": ["-y", "devcontext@latest"],
      "enabled": true,
      "env": {
        "TURSO_DATABASE_URL": "your-turso-database-url",
        "TURSO_AUTH_TOKEN": "your-turso-auth-token"
      }
    }
  }
}
```

4. Start DevContext from project root:

```bash
cd ~/Desktop/Projects/Ace_Turtle/ai_training
npx -y devcontext@latest
```

5. Use MCP tool calls in your IDE agent context.

## 5) Suggested learning flow (for this notebook)

1. Read DevContext README + installation.
2. Run `npx -y devcontext@latest` and verify service startup.
3. Configure `mcp.json` in `.cursor` or your IDE as above.
4. Use `initialize_conversation_context` once.
5. On code changes, call `update_conversation_context` and/or `record_milestone_context`.
6. Ask natural language question and call `retrieve_relevant_context`.
7. At conversation end, call `finalize_conversation_context`.

This map bridges the “understanding” goal without building a whole app.
